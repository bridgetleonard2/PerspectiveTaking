```python
import numpy as np
```

```python
data = np.load('gpt4o_topP05.npy')
```

```python
def find_last_direction_word(sentence):
    # Split the sentence into words
    words = sentence.split()
    # get rid of punctuation
    words = [word.strip('.,') for word in words]
    
    # Find the last occurrence of 'left' or 'right'
    last_left = None
    last_right = None
    
    if 'left' in words:
        last_left = len(words) - 1 - words[::-1].index('left')
        # print(last_left)
    if 'right' in words:
        last_right = len(words) - 1 - words[::-1].index('right')
        # print(last_right)
    
    # Determine the latest occurrence of 'left' or 'right'
    last_occurrence = max(last_left, last_right) if last_left is not None and last_right is not None else last_left or last_right
    
    return words[last_occurrence]

# Example usage
sentence = "The cube is on the left side of the person, and the sphere is on the right."
result = find_last_direction_word(sentence)
print(result)  # Output should be 'side'
```

```python
clean_data = []
for i in range(len(data)):
    clean_responses = []
    for j in range(len(data[i])):
        clean_responses.append(find_last_direction_word(data[i][j]))
    clean_data.append(clean_responses)
```

```python
image_names = ['left_cube_0', 'left_cube_135', 'left_cube_180', 'left_cube_225', 'left_cube_270', 'left_cube_315', 'left_cube_45', 'left_cube_90', 'right_cube_0', 'right_cube_135', 'right_cube_180', 'right_cube_225', 'right_cube_270', 'right_cube_315', 'right_cube_45', 'right_cube_90']
responses = {}
for i, name in enumerate(image_names):
    # Make all letters in data upper case
    clean_data[i] = np.char.upper(clean_data[i])  # Apply upper() to each element
    # convert clean_data[i] to a list
    clean_data[i] = clean_data[i].tolist()
    responses[name] = clean_data[i]
print(responses)
```

```python
with open("gpt4o_topP05_clean.txt", "w") as file:
    for image, response in responses.items():
        file.write(f"{image}: {response}\n")
```
